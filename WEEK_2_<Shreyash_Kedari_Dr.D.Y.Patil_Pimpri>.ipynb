### ⚠️ Why are outliers important?

Outliers may represent:

- Genuine business events
- Data entry errors
- Exceptional market conditions

Since this dataset represents real-world Tesla production and delivery information, outliers will be investigated before deciding whether to remove them.

</div>

```python
outlier_summary = []

for col in numeric_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3-Q1

    lower = Q1-1.5*IQR
    upper = Q3+1.5*IQR

    outliers = df[
        (df[col]<lower)|
        (df[col]>upper)
    ].shape[0]

    outlier_summary.append([col,outliers])

outlier_df = pd.DataFrame(
    outlier_summary,
    columns=["Feature","Outliers"]
)

outlier_df
```

#7. ⚙️ Feature Engineering

</div>

<div class="card">

## 📖 Objective

Feature Engineering is the process of creating new, meaningful variables from existing data to improve model performance.

Rather than relying only on the original features, engineered features capture hidden relationships, temporal patterns, and business insights that help machine learning algorithms make better predictions.

</div>

```python
df_fe = df.copy()
```

```python
df_fe["Quarter"] = ((df_fe["Month"] - 1) // 3) + 1

df_fe[["Month", "Quarter"]].head()
```

```python
def get_season(month):

    if month in [12,1,2]:
        return "Winter"

    elif month in [3,4,5]:
        return "Spring"

    elif month in [6,7,8]:
        return "Summer"

    return "Autumn"

df_fe["Season"] = df_fe["Month"].apply(get_season)
```

```python
df_fe["Price_per_km"] = (
    df_fe["Avg_Price_USD"] /
    df_fe["Range_km"]
)

df_fe["Production_Efficiency"] = (
    df_fe["Estimated_Deliveries"] /
    df_fe["Production_Units"]
)

df_fe["CO2_per_Delivery"] = (
    df_fe["CO2_Saved_tons"] /
    df_fe["Estimated_Deliveries"]
)

df_fe["Charging_Density"] = (
    df_fe["Charging_Stations"] /
    df_fe["Estimated_Deliveries"]
)
df_fe["Delivery_Gap"] = (
    df_fe["Production_Units"] -
    df_fe["Estimated_Deliveries"]
)

df_fe["Date"] = pd.to_datetime(
    dict(
        year=df_fe["Year"],
        month=df_fe["Month"],
        day=1
    )
)
df_fe = df_fe.sort_values("Date")

df_fe["Lag_1"] = (
    df_fe["Estimated_Deliveries"]
    .shift(1)
)

df_fe["Lag_3"] = (
    df_fe["Estimated_Deliveries"]
    .shift(3)
)
df_fe["Rolling_Mean_3"] = (
    df_fe["Estimated_Deliveries"]
    .rolling(3)
    .mean()
)

df_fe["Rolling_STD_3"] = (
    df_fe["Estimated_Deliveries"]
    .rolling(3)
    .std()
)

df_fe.fillna(method="bfill", inplace=True)
```

```python
new_features = [

    "Quarter",

    "Season",

    "Price_per_km",

    "Production_Efficiency",

    "CO2_per_Delivery",

    "Charging_Density",

    "Delivery_Gap",

    "Lag_1",

    "Lag_3",

    "Rolling_Mean_3",

    "Rolling_STD_3"

]

pd.DataFrame({
    "Engineered Features": new_features
})
```

To prevent target leakage, only features available before prediction are included in the regression model.

</div>

```python
# ============================================================
# Feature Selection
# ============================================================

target = "Estimated_Deliveries"

features = [

    "Year",

    "Month",

    "Region",

    "Model",

    "Production_Units",

    "Avg_Price_USD",

    "Battery_Capacity_kWh",

    "Range_km",

    "CO2_Saved_tons",

    "Source_Type",

    "Charging_Stations",

    "Quarter",

    "Season",

    "Price_per_km"

]

X = df_fe[features]

y = df_fe[target]
```

```python
categorical_features = X.select_dtypes(
    include="object"
).columns.tolist()

numerical_features = X.select_dtypes(
    exclude="object"
).columns.tolist()

print("Categorical Features")
print(categorical_features)

print()

print("Numerical Features")
print(numerical_features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42

)

print("Training Shape :", X_train.shape)

print("Testing Shape :", X_test.shape)
```

```python
numeric_pipeline = Pipeline([

    (

        "imputer",

        SimpleImputer(strategy="median")

    ),

    (

        "scaler",

        StandardScaler()

    )

])
```

```python
categorical_pipeline = Pipeline([

    (

        "imputer",

        SimpleImputer(strategy="most_frequent")

    ),

    (

        "encoder",

        OneHotEncoder(

            handle_unknown="ignore"

        )

    )

])
```

```python
preprocessor = ColumnTransformer(

    transformers=[

        (

            "num",

            numeric_pipeline,

            numerical_features

        ),

        (

            "cat",

            categorical_pipeline,

            categorical_features

        )

    ]

)
```

```python
display(preprocessor)
```

<div class="info">

### 🔄 Pipeline Workflow

The preprocessing pipeline is shared across all machine learning models in this project.

It performs:

- Missing value imputation
- Feature scaling for numerical features
- One-hot encoding for categorical features

The transformed dataset is then passed to different regression algorithms, ensuring consistent preprocessing and enabling a fair comparison of model performance.

</div>

```python
pipeline = Pipeline([

    (

        "preprocessor",

        preprocessor

    ),

    (

        "model",

        RandomForestRegressor(

            random_state=42

        )

    )

])
```

```python
pipeline.fit(

    X_train,

    y_train

)
```

```python
y_pred = pipeline.predict(

    X_test
)
```

```python
mae = mean_absolute_error(

    y_test,

    y_pred

)

mse = mean_squared_error(

    y_test,

    y_pred

)

rmse = np.sqrt(mse)

r2 = r2_score(

    y_test,

    y_pred
)

results = pd.DataFrame({

    "Metric":[

        "MAE",

        "MSE",

        "RMSE",

        "R² Score"

    ],

    "Value":[

        mae,

        mse,

        rmse,

        r2

    ]

})

results
```

#11. 📈 Time Series Forecasting

</div>

<div class="card">

## 📖 Objective

Time Series Forecasting is performed to estimate future Tesla vehicle deliveries based on historical delivery trends.

Unlike regression models that predict individual observations, time series models capture temporal dependencies and trends over time. In this project, monthly delivery data is aggregated and modeled using the ARIMA algorithm to forecast future deliveries.

The forecasting process includes:

- Time Series Construction
- Trend Visualization
- Stationarity Testing
- ARIMA Model Development
- Future Delivery Forecasting
- Business Interpretation

</div>

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
```

```python
# ==========================================
# Monthly Time Series Construction
# ==========================================

time_series = (
    df_fe
    .groupby("Date")["Estimated_Deliveries"]
    .sum()
    .sort_index()
)

time_series.head()
```

```python
plt.figure(figsize=(14,6))

plt.plot(
    time_series,
    linewidth=2.5,
    marker="o"
)

plt.title(
    "Historical Tesla Delivery Trend",
    fontsize=18,
    fontweight="bold"
)

plt.xlabel("Date")

plt.ylabel("Estimated Deliveries")

plt.grid(True)

plt.show()
```

<div class="info">

### 📊 Observation

The historical time series illustrates the month-wise variation in Tesla vehicle deliveries throughout the available period.

Visual inspection helps identify overall trends, fluctuations, and possible seasonal behavior before forecasting.

</div>

```python
result = adfuller(time_series)

print("="*60)
print("ADF TEST")
print("="*60)

print(f"ADF Statistic : {result[0]:.4f}")
print(f"P-value       : {result[1]:.4f}")
```

```python
if result[1] < 0.05:
    print("✅ Time series is Stationary")
else:
    print("⚠ Time series is Non-Stationary")
```

```python
arima_model = ARIMA(

    time_series,

    order=(1,1,1)

)

arima_fit = arima_model.fit()
```

```python
forecast = arima_fit.forecast(steps=12)
```

```python
future_dates = pd.date_range(

    start=time_series.index[-1] + pd.offsets.MonthBegin(1),

    periods=12,

    freq="MS"

)
```

```python
forecast_df = pd.DataFrame({

    "Date":future_dates,

    "Forecasted Deliveries":forecast.values

})

forecast_df["Forecasted Deliveries"] = (
    forecast_df["Forecasted Deliveries"]
    .round()
    .astype(int)
)

forecast_df.style\
.background_gradient(cmap="Blues")\
.hide(axis="index")
```

```python
plt.figure(figsize=(15,6))

plt.plot(

    time_series,

    label="Historical",

    linewidth=2

)

plt.plot(

    future_dates,

    forecast,

    color="red",

    marker="o",

    linewidth=2,

    label="Forecast"

)

plt.title(

    "Tesla Delivery Forecast (Next 12 Months)",

    fontsize=18,

    fontweight="bold"

)

plt.xlabel("Date")

plt.ylabel("Estimated Deliveries")

plt.legend()

plt.grid(True)

plt.show()
```
