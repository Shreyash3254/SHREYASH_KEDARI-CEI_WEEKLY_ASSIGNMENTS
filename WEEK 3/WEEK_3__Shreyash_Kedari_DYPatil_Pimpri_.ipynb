Machine Learning algorithms perform better when the data is properly prepared.

In this section, we will:

- ✅ Remove unnecessary columns
- ✅ Prepare numerical features
- ✅ Standardize the data
- ✅ Create a clean dataset for clustering

Feature scaling is particularly important because K-Means and DBSCAN are distance-based algorithms. Without scaling, features with larger values (such as GDP and Income) would dominate the clustering process.

````python
# ============================================================
# 🧹 Preparing Data for Machine Learning
# ============================================================

data = df.copy()

country_names = data["country"]

X = data.drop(columns=["country"])

print("="*60)
print("✅ Data Prepared Successfully")
print("="*60)

print("Feature Matrix Shape :", X.shape)

X.head()
````

## 📌 Why did we remove the `country` column?

The `country` column is a categorical identifier and does not contribute to distance calculations used in clustering algorithms.

However, we keep the country names separately so that we can later identify which countries belong to each cluster.

````python
# ============================================================
# ⚖️ Feature Scaling
# ============================================================

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print("="*60)
print("✅ Data Standardized Successfully")
print("="*60)

print("Scaled Dataset Shape :", X_scaled.shape)
````

````python

X_scaled_df = pd.DataFrame(
    X_scaled,
    columns=X.columns
)

X_scaled_df.head()
````

## 📌 Why Standardization?

The features in this dataset have very different ranges.

For example:

- GDP can be over 100,000.
- Inflation is generally below 100.
- Life Expectancy ranges between 30 and 90.

If we directly apply K-Means, GDP would dominate the distance calculations.

Standardization transforms every feature to have:

- Mean = 0
- Standard Deviation = 1

This ensures that all features contribute equally during clustering.

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# 🎯 STEP 6 : K-Means Clustering
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

K-Means is one of the most widely used clustering algorithms.

It partitions the data into K clusters by minimizing the distance between data points and their corresponding cluster centroids.

The first challenge is selecting the optimal number of clusters.

````python
# ============================================================
# 📈 Elbow Method
# ============================================================

inertia = []

K = range(2, 11)

for k in K:

    model = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    model.fit(X_scaled)

    inertia.append(model.inertia_)

plt.figure(figsize=(9,6))

plt.plot(
    K,
    inertia,
    marker="o",
    linewidth=3,
    markersize=8
)

plt.title(
    "Elbow Method for Optimal K",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")

plt.grid(True)

plt.show()
````

## 📌 Observation

The Elbow Method helps determine the optimal number of clusters by identifying the point where the decrease in inertia starts to slow down.

Based on the curve, **K = 3** provides a good balance between cluster compactness and model simplicity.

````python
# ============================================================
# 🎯 Training Final K-Means Model
# ============================================================

kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

kmeans_labels = kmeans.fit_predict(X_scaled)

df["KMeans_Cluster"] = kmeans_labels

print("✅ K-Means Model Trained Successfully")
````

````python
# ============================================================
# 📊 Silhouette Score
# ============================================================

score = silhouette_score(
    X_scaled,
    kmeans_labels
)

print("="*60)
print(f"Silhouette Score : {score:.3f}")
print("="*60)
````

````python
# ============================================================
# 📉 PCA Visualization
# ============================================================

pca = PCA(n_components=2)

pca_features = pca.fit_transform(X_scaled)

plt.figure(figsize=(10,7))

sns.scatterplot(
    x=pca_features[:,0],
    y=pca_features[:,1],
    hue=kmeans_labels,
    palette="viridis",
    s=100
)

plt.title(
    "K-Means Clusters (PCA Projection)",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.legend(title="Cluster")

plt.show()
````

````python
# ============================================================
# 📊 Cluster Profiling
# ============================================================

cluster_profile = df.groupby("KMeans_Cluster").mean(numeric_only=True)

display(cluster_profile)
````

````python
# ============================================================
# 📈 Cluster Profile Heatmap
# ============================================================

plt.figure(figsize=(14,8))

sns.heatmap(
    cluster_profile,
    annot=True,
    cmap="YlGnBu",
    fmt=".1f",
    linewidths=0.5
)

plt.title(
    "Average Feature Values Across Clusters",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Features")
plt.ylabel("Clusters")

plt.show()
````

The heatmap provides a visual comparison of the average socio-economic indicators across all clusters.

The cluster profile summarizes the average characteristics of each cluster.

By comparing GDP, Income, Life Expectancy, Child Mortality, and other indicators, we can understand the socio-economic profile of each group.

These clusters will later serve as labels for building supervised classification models such as Random Forest and XGBoost.

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# 🌐 STEP 7 : DBSCAN Clustering
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

Unlike K-Means, DBSCAN is a density-based clustering algorithm.

It groups closely packed observations together while identifying isolated
points as noise (outliers).

Advantages of DBSCAN:

- No need to specify the number of clusters.
- Can detect arbitrarily shaped clusters.
- Robust to noise.

````python
# ============================================================
# 🌐 DBSCAN Clustering
# ============================================================

dbscan = DBSCAN(
    eps=1.5,
    min_samples=5
)

dbscan_labels = dbscan.fit_predict(X_scaled)

df["DBSCAN_Cluster"] = dbscan_labels

print("="*60)
print("✅ DBSCAN Clustering Completed")
print("="*60)

print(df["DBSCAN_Cluster"].value_counts())
````

````python
# ============================================================
# 📉 DBSCAN Visualization using PCA
# ============================================================

plt.figure(figsize=(10,7))

sns.scatterplot(
    x=pca_features[:,0],
    y=pca_features[:,1],
    hue=dbscan_labels,
    palette="tab10",
    s=90
)

plt.title(
    "DBSCAN Clustering (PCA Projection)",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.show()
````

## 📌 Observation

Unlike K-Means, DBSCAN automatically identifies dense regions and labels
isolated observations as noise.

This makes DBSCAN suitable for datasets containing irregular cluster
shapes and potential outliers.

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# 🏷️ STEP 8 : Creating Target Labels
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

The original dataset does not contain predefined class labels.

Therefore, the K-Means cluster assignments are used as pseudo-labels to
train supervised Machine Learning models.

This approach demonstrates how unsupervised learning can be combined with
supervised learning in an end-to-end intelligence system.

````python
# ============================================================
# 🏷️ Creating Classification Dataset
# ============================================================

X = X_scaled_df

y = df["KMeans_Cluster"]

print("Feature Shape :", X.shape)
print("Target Shape :", y.shape)
````

````python
# ============================================================
# ✂️ Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Samples :", X_train.shape[0])
print("Testing Samples :", X_test.shape[0])
````

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# 🤖 STEP 9 : Classification Models
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

To compare different learning algorithms, we train multiple classifiers.

Models Used:

- Logistic Regression
- Decision Tree
- Random Forest
- XGBoost

The objective is to identify the model that best predicts the cluster
membership of unseen observations.

````python
# ============================================================
# 🤖 Training Classification Models
# ============================================================

models = {

    "Logistic Regression":
        LogisticRegression(max_iter=1000),

    "Decision Tree":
        DecisionTreeClassifier(random_state=42),

    "Random Forest":
        RandomForestClassifier(random_state=42),

    "XGBoost":
        XGBClassifier(
            random_state=42,
            eval_metric="mlogloss"
        )
}

results = []

for name, model in models.items():

    model.fit(X_train, y_train)

    prediction = model.predict(X_test)

    results.append({

        "Model": name,

        "Accuracy":
            accuracy_score(y_test, prediction),

        "Precision":
            precision_score(
                y_test,
                prediction,
                average="weighted"
            ),

        "Recall":
            recall_score(
                y_test,
                prediction,
                average="weighted"
            ),

        "F1 Score":
            f1_score(
                y_test,
                prediction,
                average="weighted"
            )
    })

results_df = pd.DataFrame(results)

results_df
````

````python
# ============================================================
# 🏆 Model Comparison
# ============================================================

results_df = results_df.sort_values(
    by="Accuracy",
    ascending=False
)

display(results_df.style.background_gradient(cmap="Greens"))
````

## 📌 Observation

The comparison table allows us to evaluate each classifier based on
multiple performance metrics.

The best-performing model will be selected for further interpretation and
feature importance analysis.

````python
# ============================================================
# 📊 Accuracy Comparison
# ============================================================

plt.figure(figsize=(9,6))

sns.barplot(
    data=results_df,
    x="Accuracy",
    y="Model",
    palette="viridis"
)

plt.xlim(0,1)

plt.title(
    "Classification Model Comparison",
    fontsize=18,
    weight="bold"
)

for index, value in enumerate(results_df["Accuracy"]):
    plt.text(
        value+0.01,
        index,
        f"{value:.3f}",
        va="center",
        fontsize=11
    )

plt.show()
````

````python
# ============================================================
# 📊 Detailed Model Performance Comparison
# ============================================================

comparison = results_df.copy()

comparison = comparison.round(3)

comparison.style\
    .background_gradient(cmap="Blues")\
    .set_caption("Classification Model Performance")
````

| 🥇 | Random Forest | 100% | Excellent performance |

### 📌 Interpretation

Both **Logistic Regression** and **Random Forest** achieved perfect classification accuracy on the generated cluster labels.

Since the labels were created using K-Means clustering on the same feature space, high predictive performance is expected. This demonstrates that the supervised models successfully learned the segmentation patterns discovered during clustering.

# 📊 STEP 10 : Confusion Matrix Analysis
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

A confusion matrix provides a detailed summary of classification performance.

Unlike accuracy alone, it shows:

- Correct Predictions
- Incorrect Predictions
- Misclassified Samples

This helps us understand how well each model performs across different classes.

````python
# ============================================================
# 📊 Confusion Matrix for All Models
# ============================================================

from sklearn.metrics import ConfusionMatrixDisplay

fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

for ax, (name, model) in zip(axes, models.items()):

    prediction = model.predict(X_test)

    cm = confusion_matrix(y_test, prediction)

    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm
    )

    disp.plot(
        cmap="Blues",
        ax=ax,
        colorbar=False
    )

    ax.set_title(name, fontsize=13)

plt.suptitle(
    "Confusion Matrices of Classification Models",
    fontsize=18,
    weight="bold"
)

plt.tight_layout()

plt.show()
````

## 📌 Observation

The confusion matrices indicate that most samples are correctly classified.

Random Forest and Logistic Regression show excellent predictive performance, while Decision Tree and XGBoost exhibit a small number of misclassifications.

Overall, all models successfully learned the cluster patterns generated by K-Means.
